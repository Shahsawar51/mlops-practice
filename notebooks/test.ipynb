```python
path = "../data/raw/sample.csv"
```

```python
import os
import pandas as pd

os.makedirs(os.path.dirname(path), exist_ok=True)

sample = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "email": ["alice@example.com", "bob@example.com", "charlie@example.com"]
})

sample.to_csv(path, index=False)
print(f"Wrote {len(sample)} rows to {path}")
```

```python
df = pd.read_csv(path)
```

```python
df
```

```python
start_id = int(df['id'].max()) + 1
n_new = 20
ids = list(range(start_id, start_id + n_new))
names = [f"User{_id}" for _id in ids]
ages = [20 + (i % 41) for i in range(n_new)]
emails = [f"user{_id}@example.com" for _id in ids]

new_rows = pd.DataFrame({"id": ids, "name": names, "age": ages, "email": emails})
df = pd.concat([df, new_rows], ignore_index=True)

df.to_csv(path, index=False)
print(f"Added {n_new} rows and saved to {path}")
```
